```python
table = np.array([
    [125, 165, 7.02, 4.19, 0.007463, 0.02086], # 10                                                                                                                                          
    [880, 1343, 6.00, 3.97, 0.004343, 0.008146], # 100 
    [6828, 12952, 4.89, 3.50, 0.00263, 0.004051],  # 1k
    [42992, 111141, 4.11, -1, 0.001713, -1] #10k
])
```
